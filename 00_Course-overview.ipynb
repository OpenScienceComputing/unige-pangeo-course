# Pangeo Pathfinder Short Course (April 18-May 5, 2024)
Emerging open-source Python-based infrastructure for scalable,data-proximate analysis and visualization of model output and remote sensing data (a.k.a the Pangeo Platform)


## Course Outline

**Day 1: April 17, 2024**

* Welcome/ workshop goals (15 min)
* [The Pangeo Platform](https://docs.google.com/presentation/d/1C_nKUSeYTtM0hdKb2yuTpCVS2VqOV4HaQwWjIxRb7i8/edit?usp=sharing) (15 min) 
* Demo: COAWST explorer (25 min)
* Demo: Create a notebook from scratch looking at data on the Unidata THREDDS Server (5 min)
 
* <u>Exercise</u>: Learn more about Xarray or Explore data (60 min)

# Deploying horizontally scaling infrastructure
  * setting up the framework for youself : 
      * Service providers: [2i2c](https://2i2c.org/), [SaturnCloud](https://saturncloud.io/), [Coiled](https://coiled.io)
      * Deploy yourself: 
          * Cloud:  deployment with [Nebari](https://www.nebari.dev/)
          * HPC/Remote machine: [Instructions for HPC or remote machine deployment](https://pangeo.io/setup_guides/hpc.html)
          * Locally: Use the [IOOS Python instructions](https://ioos.github.io/ioos_code_lab/content/ioos_installation_conda.html) to set up Python and then use our [Hytest/Pangeo environment file](https://github.com/hytest-org/hytest/blob/main/environment_set_up/nebari-users-pangeo-env.yml) instead of the IOOS environment file.
